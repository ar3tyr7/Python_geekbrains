Создадим отдельную копию данных о выжвших при крушении

        df_survived
        

```python
df_survived = data_frame.query('Survived == 1')
```

```python
df_survived
```

Псмотрим какое количество мужчин и женщин находилось на корабле в процентном соотношении.

```python
data_frame.Sex.value_counts(normalize=True).mul(100).round(2)
```

```python
count_person=data_frame.groupby("Sex", as_index = False).agg({'PassengerId':'count'}).sort_values('PassengerId', ascending=False)
```

И в количественном соотношении

```python
count_person
```

Для наглядности визуализируем полученные данные. Из данных видно, что мужчин на корабле было почти в 2-ое больше.

```python
sns.barplot(data = count_person, x = 'Sex', y ='PassengerId')
```

Посмторим на данные выживших в зависимости от пола. Процентное соотношение.

```python
df_survived.Sex.value_counts(normalize=True).mul(100).round(2)
```

Количественнное соотношение выживших в зависимости от пола.

```python
df_sex_surv=df_survived.groupby("Sex", as_index = False).agg({'PassengerId':'count'}).sort_values('PassengerId', ascending=False)
```

```python
sns.barplot(data = df_sex_surv, x = 'Sex', y ='PassengerId')
```

Проверим количество выживших в зависимости от пола и класса пассажира. Из данных видно, что первостепенно спасали женщин, класс которых выше. 

Общее количество пассажиров в срезе класса и пола.

```python
df_survived.groupby(['Pclass', 'Sex'], as_index=False).agg({'PassengerId':'count'}).sort_values('PassengerId',ascending=False)
```

Также посмотрим на процентное соотношение пассажиров корабля в срезе класса. По данным видно, что у пассажиров класса выше спаслись с большей вероятностью.

```python
data_frame.Pclass.value_counts(normalize=True).plot(kind="bar", color='purple',alpha = 0.6)
```

Посмотрим на количество пассажиров в срезе класса и пола.

```python
data_frame.groupby(['Pclass', 'Sex'], as_index=False).agg({'PassengerId':'count'}).sort_values('PassengerId',ascending=False)
```

Для полноты картины происходящего выявим средний возраст пассажира класса.

```python
data_frame.groupby('Pclass')['Age'].mean()
```

Визуализируем количество выживших и тех кому удалось спаститсь в срезе пола и потраченных средств на билет.
Можем наблюдать, что цена приобритенных билетов мужчинами не сильно влияла на их выживаемость.

```python
g = sns.FacetGrid(data_frame, hue="Survived", col="Sex", margin_titles=True,
                palette="muted",hue_kws=dict(marker=["H", "2"]))
g.map(plt.scatter, "Fare", "Age",edgecolor="w").add_legend()
plt.subplots_adjust(top=0.8)
g.fig.suptitle('Survival by Gender , Age and Fare');
```
